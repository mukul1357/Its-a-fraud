```python
X_train = pd.read_csv("/content/drive/MyDrive/dataset/DataSet/train_data.csv")
```

```python
X_test = pd.read_csv("/content/drive/MyDrive/dataset/DataSet/test_data.csv")
```

```python
X_train.head()
```

```python
X_train = X_train.drop(['Unnamed: 0'], axis=1)
```

```python
X_train.head()
```

```python
X_test.head()
```

```python
X_test = X_test.drop(['Unnamed: 0'], axis=1)
```

```python
Y_train = X_train.pop('isFraud')
```

```python
X_train.head()
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras import optimizers
```

```python
hidden_units=128
lr=0.5
hidden_layer_act='relu'
output_layer_act='sigmoid'
no_epochs=100
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train_sc = sc.fit_transform(X_train)
X_test_sc = sc.transform(X_test)
```

```python
model = Sequential()
```

```python
model.add(Dense(hidden_units, input_dim=228, activation=hidden_layer_act))
model.add(Dense(hidden_units, activation=hidden_layer_act))
model.add(Dense(hidden_units, activation=hidden_layer_act))
model.add(Dense(1, activation=output_layer_act))
```

```python
model.summary()
```

```python
model.compile(optimizer='rmsprop',
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
X_train.shape
```

```python
model.fit(X_train_sc, Y_train, epochs=no_epochs, batch_size=len(X_train_sc),  verbose=2)
```

```python
predictions = model.predict(X_train_sc)
```
